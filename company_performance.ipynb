Revenue & Profit by Category 

```python
category_summary = df.groupby('Category').agg({
    'Profit':'sum',
    'Revenue':'sum',
    'OrderID':'count'
}).reset_index()

category_summary['profit_magrin_%'] = (category_summary['Profit']/category_summary['Revenue'])*100

category_summary
```

Return Rate by Category

```python
return_by_category = df.groupby('Category').agg(
    total_orders = ('OrderID','count'),
    return_orders = ('Returned',lambda x:(x=='Yes').sum())
    ).reset_index()

return_by_category['Return_Rate_%'] = (return_by_category['return_orders']/return_by_category['total_orders'])*100

return_by_category
```

Combined Analysis

```python
combine_category_summery = category_summary.merge(return_by_category,on='Category')

combine_category_summery
```

Revenue is almost evenly distributed. 
Profit margin are very consistent. 
Return Rate is also consistent.

There is NO single problematic category.

Time Series Analysis

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['YearMonth'] = df['OrderDate'].dt.to_period('M')
```

```python
df.head()
```

```python
monthly_trend = df.groupby('YearMonth').agg(
    total_revenue = ('Revenue','sum'),
    Total_profit = ('Profit','sum'),
    Total_Orders = ('OrderID','count'),
    returned_orders = ('Returned',lambda x:(x=='Yes').sum())
).reset_index()

monthly_trend
```

```python
monthly_trend['Profit_Margin_%'] = (monthly_trend['Total_profit']/monthly_trend['total_revenue'])*100

monthly_trend['Return_Rate_%'] = (monthly_trend['returned_orders']/monthly_trend['Total_Orders']) * 100
```

```python
monthly_trend
```

Revenue remains stable with moderate volatility, profit margins are consistently strong around 30%, and return rates fluctuate between 1216% without a clear upward trend. The business appears operationally stable with manageable return-related risk.

Customer Repeat Purchase Analysis

```python
df.head()
```

```python
customer_order = df.groupby('CustomerID').agg(
    total_order = ('OrderID','count'),
    total_profit = ('Profit','sum'),
    total_revenue = ('Revenue','sum'),
    returned_orders = ('Returned',lambda x:(x=='Yes').sum())
).reset_index()

customer_order
```

```python
customer_order['Customer_Type'] = customer_order['total_order'].apply(lambda x:"Repeat" if x>1 else "One-Time")

customer_order.head()
```

```python
customer_type_summary = customer_order['Customer_Type'].value_counts()

customer_type_summary
```

```python
revenue_by_type = customer_order.groupby('Customer_Type').agg(
    Customers = ('CustomerID','count'),
    Total_Revenue = ('total_revenue','sum'),
    Total_Profit = ('total_profit','sum')
).reset_index()

revenue_by_type
```

```python
revenue_by_type["Total_Revenue"] = revenue_by_type["Total_Revenue"].map('{:,.2f}'.format)
revenue_by_type["Total_Profit"] = revenue_by_type["Total_Profit"].map('{:,.2f}'.format)

revenue_by_type
```

The companys revenue model is heavily dependent on repeat customers, who contribute more than 99% of total revenue. This indicates strong customer retention and loyalty, making repeat buyers the backbone of the business.

Average Order Value & Customer Value Analysis

```python
total_orders = len(df)

total_revenue = df['Revenue'].sum()

aov = total_revenue/total_orders

print(round(aov,2))
```

```python
aov_by_type = customer_order.copy()
```

```python
aov_by_type
```

```python
aov_by_type['AOV'] = (
    aov_by_type['total_revenue']/
    aov_by_type['total_order']
)

aov_by_type
```

```python
aov_by_type.groupby('Customer_Type')['AOV'].mean()
```

```python
customer_order.groupby('Customer_Type')['total_revenue'].mean()
```

Repeat customers do not spend more per order; however, their higher purchase frequency drives significantly greater lifetime revenue, making retention and repeat engagement the primary growth lever.
